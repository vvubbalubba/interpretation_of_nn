## ViT_Cat

```python
with open('ViT_Cat.json') as json_file:
    data = json.load(json_file)
resnet_cat = pd.DataFrame.from_dict(data)
resnet_cat.head()
```

```python
resnet_cat = resnet_cat.drop(["captioning", "annotator",
                              "annotation_id", "created_at",
                              "updated_at", "lead_time"], axis=1)
resnet_cat.loc[resnet_cat["gender"] == "Да", "gender"] = 1
resnet_cat.loc[resnet_cat["gender"] == "Нет", "gender"] = 0
resnet_cat.head()
```

```python
df = pd.read_csv('maps_for_tasks_info.csv')
arr = df.to_numpy()
idx = []
for i in range(len(arr)):
    if arr[i, 0].find('Maps_for_tasks/ViT/Cat/') != -1:
        idx.append(i)

arr = arr[idx]  
for i in range(len(arr)):
    arr[i, 0] = int(arr[i, 0].split("/")[3][0]) + 101
df = pd.DataFrame(arr, columns = ["id", "true_label", "pred_label"])
df.head()
```

```python
resnet_cat = df.join(resnet_cat.set_index('id'), on='id')
num_pos_ans = resnet_cat[resnet_cat["gender"] == 1]["gender"].count()
print(num_pos_ans / len(resnet_cat))
```
