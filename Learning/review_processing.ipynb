## ViT_Malignant

```python
with open('ViT_Malignant.json') as json_file:
    data = json.load(json_file)
    
vit_mal = pd.DataFrame.from_dict(data)

vit_mal = vit_mal.drop(["captioning", "annotator",
                              "annotation_id", "created_at",
                              "updated_at", "lead_time"], axis=1)

vit_mal.loc[vit_mal["gender"] == "Да", "gender"] = 1
vit_mal.loc[vit_mal["gender"] == "Нет", "gender"] = 0

vit_mal["id"] = (vit_mal["id"] - 1) % 100

df = pd.read_csv('maps_for_tasks_info.csv')
arr = df.to_numpy()

idx = []
for i in range(len(arr)):
    if arr[i, 0].find('Maps_for_tasks/ViT/Malignant/') != -1:
        idx.append(i)
        
arr = arr[idx]

for i in range(len(arr)):
    arr[i, 0] = int(arr[i, 0].split("/")[3][0])
    
df = pd.DataFrame(arr, columns = ["id", "true_label", "pred_label"])

vit_mal = df.join(vit_mal.set_index('id'), on='id')

num_pos_ans = vit_mal[vit_mal["gender"] == 1]["gender"].count()
print(num_pos_ans / len(vit_mal))
```
